```python
def make_concatenated_sorting(sortings: dict[str, si.BaseSorting], units: dict) -> si.NumpySorting:
    units_dict = {}
    labels = []
    
    for sorting_name in units.keys():
        for label, unit_ids in units[sorting_name].items():
            for unit_id in unit_ids:
                units_dict[len(units_dict)] = sortings[sorting_name].get_unit_spike_train(unit_id)
                labels.append(label)
    
    sf = next(iter(sortings.values())).sampling_frequency
    sorting = si.NumpySorting.from_unit_dict(units_dict, sampling_frequency=sf)
    sorting.set_property("gt_label", np.array(labels))
    
    return sorting


def remove_duplicated_units(analyzer: si.SortingAnalyzer, label: str) -> si.BaseSorting:
    sorting = analyzer.sorting
    rp = refractory_periods[label]
    C, _ = sqm.compute_refrac_period_violations(analyzer, refractory_period_ms=rp)
    assert np.all(np.array(list(C.values())) < 0.08), C  # Make sure everything is correct (C < 8%)
    
    graph = nx.Graph()
    graph.add_nodes_from(analyzer.unit_ids)
    
    delta_time = 5.0 if label == "CS" else 0.6
    redundant_units = scur.find_redundant_units(sorting, delta_time=delta_time, agreement_threshold=0.05, duplicate_threshold=0.5)
    
    for pair in redundant_units:
        graph.add_edge(*pair)
    
    for units in list(nx.connected_components(graph)):
        units = list(units)
        scores = {unit_id: len(sorting.get_unit_spike_train(unit_id)) * (1 - 3.5*C[unit_id]) for unit_id in units}
        best_unit = units[np.argmax(scores)]
        
        for unit in units:
            if unit != best_unit:
                graph.remove_node(unit)
    
    return sorting.select_units(graph.nodes)
    

if (templates_file := data_folder / "gt_templates.pkl").exists():
    gt_templates = pickle.load(open(templates_file, "rb"))
else:
    gt_templates = []
    for mouse_ID in tqdm(list(gt_units.keys())):
        folder = data_folder / mouse_ID

        with warnings.catch_warnings():  # Silence "not same version" warning.
            warnings.simplefilter("ignore")
            recording = si.load_extractor(folder / "recording.bin")
        sortings = {name: se.PhySortingExtractor(folder / "analyses" / name) for name in gt_units[mouse_ID].keys()}

        utils.Utils.sampling_frequency = recording.sampling_frequency
        utils.Utils.t_max = recording.get_num_frames()

        sorting = make_concatenated_sorting(sortings, gt_units[mouse_ID])
        labels = sorting.get_property("gt_label")

        curated_sortings = []
        for label in refractory_periods.keys():
            sub_sorting = sorting.select_units(sorting.unit_ids[labels == label])
            if sub_sorting.get_num_units() == 0:
                continue

            analyzer = si.create_sorting_analyzer(sub_sorting, recording, format="memory", sparse=False)
            sub_sorting = remove_duplicated_units(analyzer, label)
            curated_sortings.append(sub_sorting)

        # Creating the templates
        sorting = si.aggregate_units(curated_sortings)
        analyzer = si.create_sorting_analyzer(sorting, recording, format="memory", sparse=False, return_scaled=True)
        analyzer.compute({
            'random_spikes': {'method': "all"},
            'templates': {'ms_before': 15.0, 'ms_after': 25.0}
        })

        templates = analyzer.get_extension("templates").get_data(operator="average")
        filtered_templates = utils.filter(templates, [60, 12_000], axis=1)[:, 150:-150, :]
        processed_templates = MEArec.tools.pad_templates(np.swapaxes(filtered_templates, 1, 2), (1, 15), drifting=False, dtype=np.float64, verbose=False)
        processed_templates = np.swapaxes(processed_templates, 1, 2)

        for i, unit_id in enumerate(sorting.unit_ids):
            template = processed_templates[i]

            best_channel_index = np.argmax(np.max(np.abs(template), axis=0))
            amplitude_mask = np.max(np.abs(template), axis=0) >= MIN_AMPLITUDE
            distance_mask = si.get_channel_distances(recording)[best_channel_index] <= MAX_DISTANCE
            mask = amplitude_mask & distance_mask

            gt_templates.append({
                'mouse_ID': mouse_ID,
                'spike_train': sorting.get_unit_spike_train(unit_id),
                'template': processed_templates[i],
                'type': sorting.get_unit_property(unit_id, "gt_label"),
                'mask': mask,  
                'layers': recording.get_property("layer")[mask]
            })
    
    pickle.dump(gt_templates, open(templates_file, "wb"))
```

### Templates injection

Now that we have created our library of templates, we can inject a scaled-down version of them.

```python
def compute_matching_error(recording: si.BaseRecording, template_layers: np.ndarray) -> np.ndarray:
    layers = recording.get_property("layer")
    shanks = recording.get_property("shank")
    
    N = len(template_layers)
    n_combinations = recording.get_num_channels() - N + 1
    matching_error = np.empty((n_combinations, 2), dtype=np.int32)
    
    for i in range(n_combinations):
        comp = slice(i, i+N, 1)  # Comparison slice
        
        if not np.all(shanks[comp] == shanks[i]):  # If not all channels on the same shank --> highest error
            matching_error[i, :] = N
        else:
            matching_error[i, 0] = N - np.sum(layers[comp] == template_layers)
            matching_error[i, 1] = N - np.sum(layers[comp] == template_layers[::-1])
    
    return matching_error


def construct_spike_train(original_spike_train: np.ndarray, recording: si.BaseRecording, refractory_period: float) -> np.ndarray:
    if original_spike_train[-1] < recording.get_num_frames():  # Need to construct additional spikes
        ISI = np.diff(original_spike_train)
        np.random.shuffle(ISI)
        spike_train = np.concatenate((original_spike_train, original_spike_train[-1] + np.cumsum(ISI)))
    else:
        spike_train = original_spike_train
    
    spike_train = spike_train[:np.searchsorted(spike_train, recording.get_num_frames())]
    
    censored_period = int(round(refractory_period * recording.sampling_frequency * 1e-3))
    duplicate_indices = scur.find_duplicated_spikes(spike_train, censored_period, method="keep_first_iterative")
    spike_train = np.delete(spike_train, duplicate_indices)
    
    return spike_train


def construct_template(template: np.ndarray, recording: si.BaseRecording, matching_error: np.ndarray) -> np.ndarray:
    channel_indices, reverse = np.where(matching_error == np.min(matching_error))
    idx = np.random.choice(len(channel_indices))
    channel_index = channel_indices[idx]
    reverse = bool(reverse[idx])
    mask = slice(channel_index, channel_index+template.shape[1])
    
    injected_template = np.zeros([template.shape[0], recording.get_num_channels()], dtype=np.float32)
    injected_template[:, mask] = template
    if reverse:
        injected_template = injected_template[:, ::-1]
    
    return injected_template


for mouse_ID in tqdm(gt_units.keys()):
    original_recording = si.load_extractor(data_folder / mouse_ID / "recording.bin")
    injection_folder = data_folder / mouse_ID / "injection"
    
    # For each channel, which shank does it belong to?
    shanks = np.concatenate(([0], np.cumsum(np.diag(si.get_channel_distances(original_recording), 1) > 200)))
    original_recording.set_property("shank", shanks)
    
    if (injection_folder / "recording_injected.bin").exists():
        continue
    
    original_recording_f = spre.bandpass_filter(original_recording, freq_min=100, freq_max=9000, ftype="bessel", filter_order=2)
    original_noise_levels = si.get_noise_levels(original_recording_f, return_scaled=True, method="mad")
    
    injected_templates = []
    for i, template in enumerate(gt_templates):
        if template['mouse_ID'] == mouse_ID:  # Don't inject a template in the same mouse
            continue
        
        matching_error = compute_matching_error(original_recording, template['layers'])
        if np.min(matching_error) > 1:  # Could not fit template in this recording
            continue
        
        
        injected_template = construct_template(template['template'][:, template['mask']], original_recording, matching_error)
        
        # Determine scaling, but make sure that SNR >= 1
        scaling_limits = SCALING_FACTOR[template['type']]
        best_channel = np.argmax(np.max(np.abs(injected_template), axis=0))
        limit = original_noise_levels[best_channel] / np.max(np.abs(injected_templates))
        limit = min(limit, scaling_limits[1] - 0.01)
        scaling_limits[0] = max(scaling_limits[0], limit)
        injected_template *= np.random.uniform(*scaling_limits)
        
        spike_train = construct_spike_train(template['spike_train'], original_recording, refractory_periods[template['type']])
        
        injected_templates.append({
            'template_id': i,
            'template': injected_template,
            'spike_train': spike_train,
            'type': template['type'],
            'scaling': scaling
        })
    
    spike_trains = {i: injected_templates[i]['spike_train'] for i in range(len(injected_templates))}
    injected_sorting = si.NumpySorting.from_unit_dict(spike_trains, original_recording.sampling_frequency)
    injected_sorting.set_property("template_id", [t['template_id'] for t in injected_templates])
    injected_sorting.set_property("cell_type", [t['type'] for t in injected_templates])
    injected_sorting.set_property("template_scaling", [t['scaling'] for t in injected_templates])
    
    gain = original_recording.get_property("gain_to_uV")
    offset = original_recording.get_property("offset_to_uV")
    injected_templates = (np.array([t['template'] for t in injected_templates]) - offset) / gain
    injected_templates = injected_templates.round().astype(original_recording.dtype)
    injected_recording = si.inject_templates(injected_sorting, injected_templates, nbefore=300, parent_recording=original_recording)
    injected_recording = injected_recording.save(folder=injection_folder / "recording_injected.bin")
    
    analyzer = si.create_sorting_analyzer(injected_sorting, spre.gaussian_filter(injected_recording, 120, 9_000), sparse=False, return_scaled=True)
    analyzer.compute({
        'noise_levels': {},
        'random_spikes': {'max_spikes_per_unit': 5000},
        'templates': {'ms_before': 1.5, 'ms_after': 2.5}
    })
    
    # Sanity check: amplitudes match the injected amplitudes.
    amplitudes = si.get_template_amplitudes(analyzer, peak_sign="both", mode="extremum", return_scaled=True, abs_value=True)
    for unit_id, amplitude in amplitudes.items():
        template_id = injected_sorting.get_unit_property(unit_id, "template_id")
        template_amp = np.max(np.abs(gt_templates[template_id]['template']))
        template_amp *= injected_sorting.get_unit_property(unit_id, "template_scaling")
        amplitude = np.max(amplitude)
#       assert 3/4 < amplitude / template_amp < 4/3
    
    SNR = sqm.compute_snrs(analyzer, peak_sign="both")
    injected_sorting.set_property("SNR", list(SNR.values()))
    injected_sorting.save(folder=injection_folder / "gt_sorting")
```

```python
"""
The following code enables running MountainSort3 analyses from Python.

Original implementation by Victor Llobet, modified by Aurélien Wyngaard.
https://github.com/llobetv/NeurAlgo-ExtracellularAnalysis/blob/master/spikeSorting/ms3.py
"""

import csv
import json
import subprocess


mountainsort3_pipeline = """\
{"processing_server":"typhoon","pipelines":[{"spec":{"name":"main","description":"","inputs":[],"outputs":[],"parameters":[]},"steps":[{"step_type":"pipeline","pipeline_name":"synthesize","inputs":{},"outputs":{"raw":"raw","geom":"geom","waveforms_true":"waveforms_true","firings_true":"firings_true"},"parameters":{"duration":"600","samplerate":"30000"}},{"step_type":"pipeline","pipeline_name":"sort","inputs":{"raw":"raw","geom":"geom"},"outputs":{"firings_out":"firings","filt_out":"filt","pre_out":"pre"},"parameters":{"samplerate":"30000","detect_sign":""}},{"step_type":"pipeline","pipeline_name":"curate","inputs":{"pre":"pre","firings":"firings"},"outputs":{"curated_firings":"curated_firings"},"parameters":{"samplerate":"30000"}}],"input_files":[],"output_files":[]},{"spec":{"name":"synthesize","description":"","inputs":[],"outputs":[{"name":"raw"},{"name":"geom"},{"name":"waveforms_true"},{"name":"firings_true"}],"parameters":[{"name":"duration","description":"Durations of simulated dataset in seconds"},{"name":"samplerate"}]},"steps":[{"step_type":"processor","processor_name":"pyms.synthesize_random_waveforms","inputs":{},"outputs":{"waveforms_out":"waveforms_true","geometry_out":"geom"},"parameters":{"upsamplefac":"13"}},{"step_type":"processor","processor_name":"pyms.synthesize_random_firings","inputs":{},"outputs":{"firings_out":"firings_true"},"parameters":{"samplerate":"${samplerate}","duration":"${duration}"}},{"step_type":"processor","processor_name":"pyms.synthesize_timeseries","inputs":{"firings":"firings_true","waveforms":"waveforms_true"},"outputs":{"timeseries_out":"raw"},"parameters":{"duration":"${duration}","waveform_upsamplefac":"13"}},{"step_type":"processor","processor_name":"pyms.synthesize_random_firings","inputs":{},"outputs":{"firings_out":"test_firings","console_out":"test_cons"},"parameters":{"duration":"310"}}],"input_files":[],"output_files":[]},{"name":"sort","script":"/* Define the spec */\\ninputs_opt('raw','filt','pre','geom');\\noutputs('firings_out');\\noutputs_opt('filt_out','pre_out','firings_original_out');\\nparam('samplerate',30000);\\nparam('freq_min',300);\\nparam('freq_max',6000);\\nparam('freq_wid',1000);\\nparam('whiten','true');\\nparam('detect_threshold',3);\\nparam('detect_sign',0);\\nparam('adjacency_radius',-1);\\nparam('curate','false');\\nparam('consolidation_factor',0.9);\\nparam('detect_interval', 5);\\nparam('clip_size',50);\\nparam('fit_stage','true');\\n\\n_Pipeline.run=function(X) {\\n  var pp=X.parameters;\\n  \\n  var pre='pre';\\n  if (!X.hasInput('pre')) {\\n    \\n    var filt='filt';\\n    if (!X.hasInput('filt')) {\\n      if (!X.hasInput('raw')) {\\n        console.error('Missing input: raw, filt or pre');\\n        return -1;\\n      }\\n      X.runProcess('ms3.bandpass_filter',\\n                   {timeseries:'raw'},\\n                   {timeseries_out:'filt_out'},\\n                   {samplerate:pp.samplerate,freq_min:pp.freq_min,freq_max:pp.freq_max,freq_wid:pp.freq_wid}\\n                  );\\n      filt='filt_out';\\n    }\\n  \\n  \\n    if (pp.whiten=='true') {\\n      X.runProcess('ms3.whiten',\\n                   {timeseries:filt},\\n                   {timeseries_out:'pre_out'},\\n                   {}\\n                  );\\n    }\\n    else {\\n      X.runProcess('pyms.normalize_channels',\\n                   {timeseries:'filt'},\\n                   {timeseries_out:'pre_out'},\\n                   {}\\n                  );\\n    }\\n    pre='pre_out';\\n  }\\n  \\n  \\n  var curate=(pp.curate=='true');\\n  var firings1='firings_out';\\n  if (curate) firings1='firings_original_out';\\n  \\n  var p={\\n    detect_threshold:pp.detect_threshold,\\n    detect_sign:pp.detect_sign,\\n    adjacency_radius:pp.adjacency_radius\\n,\\n    clip_size:pp.clip_size\\n,\\n    detect_interval:pp.detect_interval\\n,\\n    consolidation_factor:pp.consolidation_factor\\n,\\n    fit_stage:pp.fit_stage\\n  };\\n  var inputs={timeseries:pre};\\n  if (X.hasInput('geom')) {\\n    inputs.geom='geom';\\n  }\\n  X.runProcess('mountainsortalg.ms3',\\n               inputs,\\n               {firings_out:firings1},\\n               p);\\n  \\n  if (curate) {\\n    X.runPipeline('curate',\\n               {pre:pre,firings:firings1},\\n               {curated_firings:'firings_out'},\\n               {samplerate:pp.samplerate});\\n  }\\n               \\n};\\n\\n/////////////////////////////////////////////////////////////////////\\n\\n\\nfunction param(str,val) {\\n      if (val===undefined) {\\n        _Pipeline.spec.parameters.push({name:str});\\n      }\\n      else {\\n        _Pipeline.spec.parameters.push({name:str,optional:true,default_value:val});\\n      }\\n}\\n                \\nfunction inputs(str1,str2,str3,str4) {\\n  if (str1) _Pipeline.spec.inputs.push({name:str1});\\n  if (str2) _Pipeline.spec.inputs.push({name:str2});\\n  if (str3) _Pipeline.spec.inputs.push({name:str3});\\n  if (str4) _Pipeline.spec.inputs.push({name:str4});\\n}\\n\\nfunction inputs_opt(str1,str2,str3,str4) {\\n  if (str1) _Pipeline.spec.inputs.push({name:str1,optional:true});\\n  if (str2) _Pipeline.spec.inputs.push({name:str2,optional:true});\\n  if (str3) _Pipeline.spec.inputs.push({name:str3,optional:true});\\n  if (str4) _Pipeline.spec.inputs.push({name:str4,optional:true});\\n}\\n\\nfunction outputs(str1,str2,str3,str4) {\\n  if (str1) _Pipeline.spec.outputs.push({name:str1});\\n  if (str2) _Pipeline.spec.outputs.push({name:str2});\\n  if (str3) _Pipeline.spec.outputs.push({name:str3});\\n  if (str4) _Pipeline.spec.outputs.push({name:str4});\\n}\\n\\nfunction outputs_opt(str1,str2,str3,str4) {\\n  if (str1) _Pipeline.spec.outputs.push({name:str1,optional:true});\\n  if (str2) _Pipeline.spec.outputs.push({name:str2,optional:true});\\n  if (str3) _Pipeline.spec.outputs.push({name:str3,optional:true});\\n  if (str4) _Pipeline.spec.outputs.push({name:str4,optional:true});\\n}","steps":[],"spec":{"name":"","description":"","inputs":[],"outputs":[],"parameters":[]},"export":true},{"spec":{"name":"curate","description":"","inputs":[{"name":"pre"},{"name":"firings"}],"outputs":[{"name":"curated_firings"}],"parameters":[{"name":"samplerate"}]},"steps":[{"step_type":"processor","processor_name":"ms3.cluster_metrics","inputs":{"timeseries":"pre","firings":"firings"},"outputs":{"cluster_metrics_out":"metrics1"},"parameters":{"samplerate":"${samplerate}"}},{"step_type":"processor","processor_name":"ms3.isolation_metrics","inputs":{"timeseries":"pre","firings":"firings"},"outputs":{"metrics_out":"metrics2"},"parameters":{"compute_bursting_parents":"true"}},{"step_type":"processor","processor_name":"ms3.combine_cluster_metrics","inputs":{"metrics_list":["metrics1","metrics2"]},"outputs":{"metrics_out":"metrics"},"parameters":{}},{"step_type":"processor","processor_name":"pyms.create_label_map","inputs":{"metrics":"metrics"},"outputs":{"label_map_out":"label_map"},"parameters":{}},{"step_type":"processor","processor_name":"pyms.apply_label_map","inputs":{"firings":"firings","label_map":"label_map"},"outputs":{"firings_out":"curated_firings"},"parameters":{}}],"input_files":[],"output_files":[]}],"input_files":[],"output_files":[],"jobs":[]}\
"""


def recording_to_mda(recording: si.BaseRecording, tmp_folder: str) -> None:
	"""
	Saves the recording traces in mda format.
	Takes in the traces [n_samples, n_channels] and writes it in 1D in C format
	(i.e. t1 c1, t1 c2, t1 c3, ..., t1 cn, t2 c1, ...).

	mda format: (http://mountainsort.readthedocs.io/en/latest/mda_file_format.html)
	array[int32]:
		- 1st: negative integer about data format:
			-1 = complex float32
			-2 = byte
			-3 = float32
			-4 = int16
			-5 = int32
			-6 = uint32
			-7 = double
			-8 = uint32
		- 2nd: number of byte of data type
		- 3rd: number of dimensions (typically 2, time and channels)
		- 4th: number of channels
		- 5th: number of samples in time
	"""

	filename = f"{tmp_folder}/data.mda"

	mda_header = np.empty(5, dtype=np.int32)
	mda_header[0] = -4  # int16
	mda_header[1] = 2  # 2 bytes int int16
	mda_header[2] = 2  # 2 dimensions
	mda_header[3] = recording.get_num_channels()
	mda_header[4] = recording.get_num_frames()

	assert recording.dtype == np.int16
	data = recording.get_traces()

	with open(filename, 'wb') as file:
		mda_header.tofile(file)
	with open(filename, 'ab') as file:  # 'ab' means append rather than starting from scratch.
		data.tofile(file)


def save_geometry(recording: si.BaseRecording, tmp_folder: str) -> None:
	"""
	Takes the geometry of a recording object, and save it in the format that ms3 expects.
	"""

	geometry = recording.get_property("location")

	with open(f"{tmp_folder}/geom.tsv", 'w') as file:
		writer = csv.writer(file, delimiter='\t', lineterminator='\n')
		for channel_geom in geometry:
			writer.writerow(channel_geom)


def launch_ms3(recording: si.BaseRecording, output_folder: str, ms3_params: dict, tmp_folder: str = "/mnt/hdd/tmp/mountainlab/ms3") -> se.MdaSortingExtractor:
	"""
	Runs a mountainsort3 analysis and return the sorting object.
	"""

	Path(tmp_folder).mkdir(parents=True, exist_ok=True)

	recording_to_mda(recording, tmp_folder)
	save_geometry(recording, tmp_folder)
	with open(f"{tmp_folder}/params.json", 'w') as file:
		json.dump(ms3_params, file)
	with open(f"{tmp_folder}/mountainsort3.mlp", 'w') as file:
		file.write(mountainsort3_pipeline)

	bash_command = "mlp-run mountainsort3.mlp sort --raw=data.mda --geom=geom.tsv --firings_out=firings.mda --_params=params.json"
	process = subprocess.Popen(bash_command.split(), stdout=subprocess.PIPE, cwd=tmp_folder)
	output, error = process.communicate()

	Path(output_folder).mkdir(parents=True, exist_ok=True)

	with open(f"{output_folder}/logs.txt", 'w') as file:
		file.write(output.decode(encoding="utf-8"))
	if error is not None:
		with open(f"{output_folder}/error.txt", 'w') as file:
			file.write(error.decode(encoding="utf-8"))

	sorting = se.MdaSortingExtractor(f"{tmp_folder}/firings.mda", sampling_frequency=recording.sampling_frequency)

	shutil.rmtree(tmp_folder)
	shutil.rmtree("/mnt/raid0/tmp/mountainsort/mountainlab/prvbucket")

	return sorting
```

```python
all_df = pd.DataFrame()
n_wrong = {analysis_name: {} for analysis_name in all_analyses_name}

for mouse_ID in tqdm(gt_units.keys()):
    injection_folder = data_folder / mouse_ID / "injection"
    
    with warnings.catch_warnings():  # Silence "not same version" warning.
        warnings.simplefilter("ignore")
        injected_recording = si.load_extractor(injection_folder / "recording_injected.bin")
    gt_sorting = si.load_extractor(injection_folder / "gt_sorting")
    n_spikes_gt = gt_sorting.count_num_spikes_per_unit(outputs="array")
    
    df = pd.DataFrame(index=gt_sorting.unit_ids, columns=['mouse_ID', 'neuron_ID', 'type', 'scale', 'SNR'] + analyses_name, dtype=float)
    df['mouse_ID'] = mouse_ID
    df['neuron_ID'] = gt_sorting.unit_ids
    df['type'] = gt_sorting.get_property("cell_type")
    df['scale'] = gt_sorting.get_property("template_scaling").round(3)
    df['SNR'] = gt_sorting.get_property("SNR").round(2)
    mask_CS = (df['type'] == "CS").to_numpy()
    
    for analysis_name in all_analyses_name:
        sorting = si.load_extractor(injection_folder / "analyses" / analysis_name / "sorting")
        n_spikes = sorting.count_num_spikes_per_unit(outputs="array")
        
        cross_shifts = utils.compute_cross_shift_from_vector(gt_sorting.to_spike_vector(), sorting.to_spike_vector(), 45)
        tp = utils.compute_coincidence_matrix_from_vector(gt_sorting.to_spike_vector(), sorting.to_spike_vector(), 6, cross_shifts)
        tp[mask_CS, :] = utils.compute_coincidence_matrix_from_vector(gt_sorting.to_spike_vector(), sorting.to_spike_vector(), 15, cross_shifts)[mask_CS, :]
        fn = n_spikes_gt[:, None] - tp
        fp = n_spikes[None, :] - tp
        agreement_matrix = tp / (tp + fp + fn)
        precision_matrix = tp / (tp + fp)
        recall_matrix = tp / (tp + fn)
        
        accuracy_per_neuron = np.max(agreement_matrix, axis=1)
        accuracy_per_neuron[accuracy_per_neuron < 1/3] = 0.0
        df[analysis_name] = accuracy_per_neuron
        
        indices = np.where((recall_matrix > 1/3) & (agreement_matrix < 0.5))
        for i, j in zip(*indices):
            if precision_matrix[i, j] < 0.5:
                neuron_id = gt_sorting.unit_ids[i]
                cell_type = gt_sorting.get_unit_property(neuron_id, "cell_type")
                if cell_type in n_wrong[analysis_name]:
                    n_wrong[analysis_name][cell_type] += 1
                else:
                    n_wrong[analysis_name][cell_type] = 1
    
    all_df = pd.concat((all_df, df), axis="index", ignore_index=True)

all_df.sort_values(["type", "scale"], inplace=True)

with open(data_folder / "analyzed_df.pkl", "wb+") as file:
    pickle.dump(all_df, file)
with open(data_folder / "garbage_units.pkl", "wb+") as file:
    pickle.dump(n_wrong, file)

all_df
```
