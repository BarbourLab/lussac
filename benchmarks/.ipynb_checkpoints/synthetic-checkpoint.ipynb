### Generating the dataset with `MEArec`

Here, we generate a synthetic dataset with 3,000 neurons to mimic the density of neurons found in the cerebral cortex (the density is around 50,000-150,000 neurons/mm^3: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3800983/ ; https://doi.org/10.3389%2Ffnana.2013.00035).

The reasoning behind is to generate neurons with different SNR as close to a real-life scenario as possible, while also generating noise with neurons having a really low SNR (< 1.5).

The dataset is 1 hour long with drift, and the recording probe uses the layout of the Neuropixels 1, but with 128 channels.

```python
n_cells = 3_000

recording_params = mr.get_default_recordings_params()
templates_params = mr.get_default_templates_params()
cell_models_folder = mr.get_default_cell_models_folder()

# Set the templates parameters
templates_filename = data_folder / "synthetic_templates.h5"
templates_params['probe'] = "Neuropixels-128"
templates_params['dt'] = 1/30			# 30 kHz
templates_params['min_amp'] = 0			# Output all templates, regardless of their min amplitude (in µV).
templates_params['xlim'] = [8, 160]		# Setting the limits for the neurons location (in µm for x, y and z axes).
templates_params['ylim'] = [-160, 160]
templates_params['zlim'] = [-520, 800]
templates_params['x_distr'] = "uniform"
templates_params['n'] = 2000
templates_params['drifting'] = WITH_DRIFT
templates_params['drift_xlim'] = [-2, 2]  # Set a very small maximum drift in XY, and ~100 µm max drift in Z.
templates_params['drift_ylim'] = [-2, 2]
templates_params['drift_zlim'] = [98, 102]
templates_params['drift_steps'] = 60

# Set the recording parameters
recording_filename = data_folder / "synthetic_recording.h5"
recording_params['recordings']['dtype'] = "float32"
recording_params['recordings']['bursting'] = False		# No bursting amplitude modulation (takes too much RAM).
recording_params['recordings']['n_bursting'] = 0		# No bursting unit.
recording_params['recordings']['modulation'] = "template"
recording_params['recordings']['chunk_duration'] = 2	# in seconds.
recording_params['recordings']['noise_level'] = 10		# Add a Gaussian with SD 10 µV for noise.
recording_params['recordings']['noise_mode'] = "uncorrelated"
recording_params['recordings']['drifting'] = WITH_DRIFT
recording_params['recordings']['filter'] = False		# Output the unfiltered recording.
recording_params['recordings']['overlap'] = False
recording_params['recordings']['extract_waveforms'] = False
recording_params['spiketrains']['duration'] = 3_600		# in seconds.
recording_params['spiketrains']['n_exc'] = (n_cells * 4) // 5	# 20% of the neurons are inhibitory in the cerebral cortex.
recording_params['spiketrains']['n_inh'] = n_cells - recording_params['spiketrains']['n_exc']
recording_params['templates']['min_dist'] = 18			# At least 18 µm between cells center.
recording_params['templates']['min_amp'] = 1			# Only keep cell if its signal is at least 1 µV.
recording_params['templates']['max_amp'] = 350			# In µV.
drift_dicts = [mr.get_default_drift_dict(), mr.get_default_drift_dict()]  # Slow and fast drifts
drift_dicts[0]['drift_mode_probe'] = "rigid"
drift_dicts[0]['drift_mode_speed'] = "slow"
drift_dicts[0]['slow_drift_velocity'] = 2  # 2 µm/min = 120 µm/h
drift_dicts[0]['slow_drift_amplitude'] = 55  # In µm.
drift_dicts[0]['slow_drift_waveform'] = "sine"
drift_dicts[1]['drift_mode_speed'] = "fast"
drift_dicts[1]['fast_drift_period'] = 600  # Fast drift event every 10 minutes.
drift_dicts[1]['fast_drift_min_jump'] = 5  # Fast drift events between 5 and 20 µm
drift_dicts[1]['fast_drift_max_jump'] = 20

# Seed to that people can re-generate the same data file.
recording_params['seeds']['spiketrains'] = 42
recording_params['seeds']['templates'] = 42
recording_params['seeds']['convolution'] = 42
recording_params['seeds']['noise'] = 42
templates_params['seed'] = 42

# Print the density of neurons
volume_um = (templates_params['xlim'][1] - templates_params['xlim'][0]) * \
			(templates_params['ylim'][1] - templates_params['ylim'][0]) * \
			(templates_params['zlim'][1] - templates_params['zlim'][0])
volume_mm = volume_um * 1e-9
print(f"{n_cells/volume_mm:.0f} neurons / mm^3")

# Generating the templates file (if it doesn't exist)
if not templates_filename.exists():
	print("\n\nGenerating templates ...")
	(tmp_folder / "templates").mkdir(parents=True, exist_ok=True)
	templates_generator = mr.gen_templates(cell_models_folder=cell_models_folder, params=templates_params, templates_tmp_folder=tmp_folder / "templates", verbose=True, n_jobs=3 if WITH_DRIFT else 8)
	mr.save_template_generator(tempgen=templates_generator, filename=templates_filename)
	del templates_generator  # Only the file is necessary. Not the variable.

# Generating the recording file (if it doesn't exist)
if not os.path.exists(recording_filename):
	print("\n\nGenerating recording ...")
	(tmp_folder / "recording").mkdir(parents=True, exist_ok=True)
	recording_generator = mr.gen_recordings(templates=templates_filename, params=recording_params, tmp_folder=tmp_folder / "recording", drift_dicts=drift_dicts, verbose=True, n_jobs=2)
	mr.save_recording_generator(recording_generator, recording_filename, include_spike_traces=False)
	del recording_generator  # This will delete the heavy temp files.

recording_generator = mr.load_recordings(recording_filename)

# Save the recording to binary format for easier handling and loading
if not (rec_bin := data_folder / "recording.bin").exists():
    print("Exporting recording to binary format ...")
    
    # Gain of 2.34 for Neuropixels 1.0: https://github.com/cortex-lab/neuropixels/wiki/Gain_settings
    recording = spre.scale(se.MEArecRecordingExtractor(recording_filename), gain=1/2.34).astype(np.int16, round=True)
    recording.set_property("gain_to_uV", [2.34]*recording.get_num_channels())
    if TETRODES:
        recording = recording.channel_slice(recording.channel_ids[TETRODES_CHANNELS])
    recording = recording.save(folder=rec_bin)

    recording_f = spre.gaussian_filter(recording, freq_min=150, freq_max=7000)
    recording_f = recording_f.save(folder=data_folder / "recording_filtered.bin")

    recording_corrected = spre.correct_motion(recording_f, preset="nonrigid_fast_and_accurate", folder=data_folder / "motion_info")
    recording_corrected = recording_corrected.save(folder=data_folder / "recording_corrected.bin")
    motion_info = spre.load_motion_info(data_folder / "motion_info")
else:
    print("Loading recording from binary format ...")
    recording = si.load_extractor(rec_bin)
    recording_f = si.load_extractor(data_folder / "recording_filtered.bin")
    recording_corrected = si.load_extractor(data_folder / "recording_corrected.bin")
    motion_info = spre.load_motion_info(data_folder / "motion_info")

# Save the ground truth sorting with original templates
if not (gt_folder := analyses_folder / "ground_truth" / "sorting").exists():
    print("Exporting ground truth to numpy folder ...")

    gt_sorting = se.MEArecSortingExtractor(recording_filename)
    gt_sorting.annotate(name="ground_truth")
    
    templates_max = np.max(np.abs(recording_generator.original_templates), axis=-1)  # in µV.
    if WITH_DRIFT:
        templates_max = np.max(templates_max, axis=1)
    if TETRODES:
        templates_max = templates_max[:, TETRODES_CHANNELS]
    gt_sorting.set_property("max_amplitude", np.max(templates_max, axis=1))

    best_channels = np.argmax(templates_max, axis=1)
    noise_levels = si.get_noise_levels(recording_f, return_scaled=True, num_chunks_per_segment=100)
    SNR = np.max(templates_max, axis=1) / noise_levels[best_channels]
    gt_sorting.set_property("SNR", SNR)
    
    gt_sorting = gt_sorting.save(folder=gt_folder)
else:
    print("Loading ground truth ...")
    gt_sorting = si.load_extractor(gt_folder)
```

### Plotting information about the neurons

Plotting the (initial) position of each neuron in space relative to the probe, as well as the relationship between the neuron's max amplitude against how far they are

```python
channel_pos = np.array(recording_generator.channel_positions)     # shape [n_channels, 3], µm
templates_pos = np.array(recording_generator.template_locations)  # shape [n_neurons, 3], µm
if WITH_DRIFT:
    templates_pos = templates_pos[:, 0, :]
if TETRODES:
    channel_pos = channel_pos[TETRODES_CHANNELS]
mask_excitatory = [any([cat in a.split('_')[1] for cat in recording_generator.info['cell_types']['excitatory']]) for a in recording_generator.template_celltypes]

fig = go.Figure().set_subplots(rows=2, cols=2)

# Plotting distribution of cells
for i in range(3):  # i=0 -> XY ; i=1 => XZ ; i=2 => YZ
    row = 1 + i//2
    col = 1 + i%2
    plan = "XY" if i==0 else "XZ" if i==1 else "YZ"

    fig.add_trace(go.Scatter(
        x=templates_pos[:, i//2],
        y=templates_pos[:, (i+3)//2],
        mode="markers",
        marker=dict(
            color="LightCoral",
            size=5
        ),
        name=f"Neurons pos ({plan})",
        legendgroup="neurons_pos",
        text=gt_sorting.unit_ids
    ), row=row, col=col)
    
    fig.add_trace(go.Scatter(
        x=channel_pos[:, i//2],
        y=channel_pos[:, (i+3)//2],
        mode="markers",
        marker=dict(
            color="LightSeaGreen",
            size=8
        ),
        name=f"Channels pos ({plan})",
        legendgroup="channel_pos",
        text=recording.channel_ids
    ), row=row, col=col)

    fig.update_xaxes(title_text=f"{plan[0]} axis (µm)", row=row, col=col)
    fig.update_yaxes(title_text=f"{plan[1]} axis (µm)", row=row, col=col)

# Plotting max amplitude VS distance to nearest channel
amplitudes = gt_sorting.get_property("max_amplitude")  # shape [n_neurons], µV
distance_to_nearest = np.min(np.sqrt(np.sum((templates_pos[:, None, :] - channel_pos[None, :, :]) ** 2, axis=2)), axis=1)

fig.add_trace(go.Scatter(
    x=distance_to_nearest,
    y=amplitudes,
    mode="markers",
    marker_color="LightCoral",
    name="Neurons' amplitude",
    text=gt_sorting.unit_ids
), row=2, col=2)

fig.update_xaxes(title_text="Distance to nearest channel (µm)", row=2, col=2)
fig.update_yaxes(title_text="Highest amplitude (µV)", row=2, col=2)
fig.update_layout(title_text="Neurons position and amplitude")

fig.show(renderer="svg")
```

Plotting the electrode drift, as well as the motion estimation from SpikeInterface (non-rigid, mean with min/max)

```python
if WITH_DRIFT:
    dt = recording_generator.drift_list[0]['drift_fs'] * recording.sampling_frequency * 1e-4
else:
    dt = recording.sampling_frequency / 100
t_axis = np.arange(0, recording.get_num_frames(), dt) / recording.sampling_frequency
gt_drift = np.zeros(t_axis.shape, dtype=np.float32)

if WITH_DRIFT:
    for drift in recording_generator.drift_list:
        gt_drift += np.array(drift['drift_vector_um'])

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=t_axis,
    y=gt_drift,
    mode="lines",
    marker_color="CornflowerBlue",
    name="GT drift"
))
fig.add_trace(go.Scatter(
    x=motion_info['temporal_bins'],
    y=np.mean(motion_info['motion'], axis=1) - np.median(motion_info['motion']) + np.median(gt_drift),
    mode="lines",
    marker_color="Crimson",
    name="Computed drift",
    legendgroup="detected_drift"
))
fig.add_trace(go.Scatter(
    x=np.concatenate((motion_info['temporal_bins'], motion_info['temporal_bins'][::-1])),
    y=np.concatenate((np.min(motion_info['motion'], axis=1), np.max(motion_info['motion'][::-1], axis=1))) - np.median(motion_info['motion']) + np.median(gt_drift),
    fill="toself",
    fillcolor="rgba(220, 20, 60, 0.4)",
    line_color="rgba(220, 20, 60, 0.4)",
    hoverinfo="skip",
    legendgroup="detected_drift",
    showlegend=False
))

fig.update_xaxes(title_text="Time (s)")
fig.update_yaxes(title_text="Drift in z-axis (µm)")

fig.show(renderer="svg")
```

### Running spike-sorting analyses

Here, we run the spike-sorting analyses on the newly generated recording.

The algorithms are ran using the dockerized sorters of SpikeInterface.

```python
# Maps name of the analysis to the parameters for the run.
analyses_params = {
	'ks2_default': {
		'sorter_name': "kilosort2",
		'NT': 5 * 64 * 1024 + 64  # Quintuple the default value (otherwise it crashes).
	},
	'ks2_pj8-3': {
		'sorter_name': "kilosort2",
		'projection_threshold': [8, 3],
		'NT': 5 * 64 * 1024 + 64,  # Qunintuple the default value (otherwise it crashes).
		'freq_min': 40,
		'recording': {
			'filter': {
				'band': [200, 6_000],
				'filter_order': 2,
				'ftype': "bessel",
				'margin_ms': 10.0
			}
		}
	},
# 	'ks2_pj6-2.5': {
# 		'sorter_name': "kilosort2",
# 		'projection_threshold': [6, 2.5],
# 		'NT': 5 * 64 * 1024 + 64,  # Qunintuple the default value (otherwise it crashes).
# 		'freq_min': 40,
# 		'recording': {
# 			'gaussian_filter': {
# 				'freq_min': 180,
# 				'freq_max': 5_000
# 			}
# 		}
# 	},
	'ks2.5_default': {
		'sorter_name': "kilosort2_5",
		'do_correction': not TETRODES  # No drift correction on tetrodes recording
	},
	'ks2.5_pj7-3': {
		'sorter_name': "kilosort2_5",
		'do_correction': not TETRODES,  # No drift correction on tetrodes recording
		'projection_threshold': [7, 3],
		'freq_min': 40,
		'recording': {
			'gaussian_filter': {
				'freq_min': 150,
				'freq_max': 5_000
			}
		}
	},
	'ks3_default': {
		'sorter_name': "kilosort3",
		'do_correction': not TETRODES  # No drift correction on tetrodes recording
	},
	'ks3_pj7-3': {
		'sorter_name': "kilosort3",
		'do_correction': not TETRODES,  # No drift correction on tetrodes recording
		'projection_threshold': [7, 3]
	},
    'ks4_default': {
        'sorter_name': "kilosort4",
		'do_correction': not TETRODES,  # No drift correction on tetrodes recording
        'duplicate_spike_bins': 12  # 0.4 ms
    },
    'ks4_low-thresh': {
        'sorter_name': "kilosort4",
		'do_correction': not TETRODES,  # No drift correction on tetrodes recording
        'duplicate_spike_bins': 12,  # 0.4 ms
        'Th_universal': 7,
        'Th_learned': 6,
		'recording': {
			'gaussian_filter': {  # Add low-pass filter.
				'freq_min': None,
				'freq_max': 5_000
			}
		}
        
    },
	'sc_default': {
		'sorter_name': "spykingcircus",
        'filter': False,  # Apply the filter ourself to apply drift correction
        'recording': {
            'bandpass_filter': {
                'freq_min': 300.0,
                'freq_max': 0.95 * recording.sampling_frequency / 2,
                'filter_order': 3,
                'ftype': "butter"
            },
            'correct_motion': {
                'preset': "nonrigid_fast_and_accurate",
                'progress_bar': False
            }
        }
	},
	'sc2_default': {
		'sorter_name': "spykingcircus2"
	},
	'tdc_default': {
		'sorter_name': "tridesclous",
        'freq_min': 40.0,  # Apply the filter ourself to apply drift correction
        'freq_max': 14500.0,
        'recording': {
            'bandpass_filter': {
                'freq_min': 400.0,
                'freq_max': 5000.0,
                'filter_order': 5,
                'ftype': "butter"
            },
            'correct_motion': {
                'preset': "nonrigid_fast_and_accurate",
                'progress_bar': False
            }
        }
	},
# 	'ms5_default': {
# 		'sorter_name': "mountainsort5",
# 		'scheme': "2",
#         'filter': False,  # Apply the filter ourself to apply drift correction
#         'recording': {
#             'bandpass_filter': {
#                 'freq_min': 300.0,
#                 'freq_max': 6000.0,
#                 'dtype': "float32"
#             },
#             'correct_motion': {
#                 'preset': "nonrigid_fast_and_accurate",
#                 'progress_bar': False
#             }
#         }
# 	}
}

times = np.memmap(analyses_folder / "analyses_time.npy", dtype=np.float32, shape=(len(analyses_params),),
                  mode="r+" if (analyses_folder / "analyses_time.npy").exists() else "w+")

# Run the spike-sorting algorithms
i = 0
for name, params in list(analyses_params.items()):
	analysis_folder = analyses_folder / name
    
	# Remove analyses that don't work with tetrodes.
	if TETRODES and (name.startswith("ks2_5") or name.startswith("ks3") or name.startswith("ks4")):
		del analyses_params[name]
		continue
    
	if analysis_folder.exists():
		try:  # Try loading the sorting. If it works, then analysis already ran --> skip it.
			sorting = si.load_extractor(analysis_folder / "sorting")
			print(f"Analysis '{name}' already ran and successful")
			continue
		except:  # Probably the folder is the result of an analysis that crashed --> delete it.
			shutil.rmtree(analysis_folder)

	print(f"Running '{name}'")
	t_start = time.perf_counter()

	recording_tmp = recording
	if 'recording' in params:
		for func_name, p in params['recording'].items():
		 	func = getattr(spre, func_name)
		 	recording_tmp = func(recording_tmp, **p)

	params.update({
		'recording': recording_tmp,
		'folder': analysis_folder,
		'delete_output_folder': True,
		'singularity_image': params['sorter_name'] != "spykingcircus2",
		'delete_container_files': True
	})
	if params['singularity_image']:
		params['extra_requirements'] = ["numpy==1.26"]  # Don't install Numpy 2.0! (incompatible)'

	if name.startswith("ks2") or name.startswith("ks3"):  # Delete unnecessary heavy files.
		params.update({
			'delete_tmp_files': True,
			'delete_recording_dat': True,
		})

	sorting = ss.run_sorter(**params)
	t_end = time.perf_counter()
	times[i] = t_end - t_start
    
	sorting._recording = None  # Remove warning
	sorting.save(folder=analysis_folder / "sorting")
	i += 1
```

```python
# Defining the parameters for automatic classifying
COINCIDENT_WINDOW = 6			 # If two spikes are <= time samples appart, they are considered coincident.
SMALL_UNITS_MAX_FR = 0.4		 # Minimum mean firing rate (in Hz, below that -> small unit).
MAX_CONTAMINATION = 0.25		 # Maximum contamination (below that -> contaminated unit).
MIN_FRAC_SPIKES_NOISY = 0.5		 # Minimum fraction of events that match to spikes (below that -> noisy unit).
RECALL_2ND_BEST_MERGED = 1/3	 # If a unit recalls more than 1/3 spikes from 2 different neurons -> merged unit.
MUA_MAX_PRECISION = 0.5			 # Minimum precision on at least one neuron (below that -> mua).
MATCH_ACC = [0.95, 0.8, 0.5]	 # Thresholds for splitting the matched units.
SHARED_FRAC_REDUNDANT = 0.5		 # Minimum fraction of spikes between units to be considered redundant.
MIN_PRECISION_GOOD_NEURON = 0.8	 # Minimum precision for a unit to be chosen for creating the theoretical maximum.

MATCH_LABELS = [f"acc >= {MATCH_ACC[0]:.0%}"]
for i in range(len(MATCH_ACC) - 1):
	MATCH_LABELS.append(f"{MATCH_ACC[i]:.0%} > acc >= {MATCH_ACC[i+1]:.0%}")
MATCH_LABELS.append(f"{MATCH_ACC[-1]:.0%} > acc")

recording_f = si.load_extractor(data_folder / "recording_filtered.bin")
params_wvf = {  # Parameters for extracting waveforms
	'ms_before': 3.0,
	'ms_after': 5.0,
	'max_spikes_per_unit': 20_000
}


def compute_coincidence_matrix(sorting1: si.BaseSorting, sorting2: si.BaseSorting) -> np.ndarray:
    """
    Computes the number of coincident spikes between each pair (unit1, unit2)
    where unit1 is in sorting1, and unit2 in sorting2.
    """

    name1 = sorting1.get_annotation("name")
    name2 = sorting2.get_annotation("name")
    assert name1 is not None and name2 is not None
    file = tmp_folder / "coincidence_matrices" / f"{name1}_with_{name2}.npy"

    if file.exists():
        coincidence_matrix = np.load(file)
    else:
        spike_vector1 = np.array(sorting1.to_spike_vector())
        spike_vector2 = np.array(sorting2.to_spike_vector())
        cross_shifts = lussac.utils.compute_cross_shift_from_vector(spike_vector1, spike_vector2, max_shift=30)
        coincidence_matrix = lussac.utils.compute_coincidence_matrix_from_vector(spike_vector1, spike_vector2, COINCIDENT_WINDOW, cross_shifts).astype(np.int32)
    
        file.parent.mkdir(exist_ok=True, parents=True)
        np.save(file, coincidence_matrix)

    return coincidence_matrix


def save_gt_properties(sorting: si.NumpySorting, path: Path):
    assert all(key in sorting.get_property_keys() for key in ["gt_label", "gt_neuron", "gt_accuracy", "gt_precision", "gt_recall"])
    
    np.save(path / "gt_label.npy", sorting.get_property('gt_label'))
    np.save(path / "gt_neuron.npy", sorting.get_property('gt_neuron'))
    np.save(path / "gt_accuracy.npy", sorting.get_property('gt_accuracy'))
    np.save(path / "gt_precision.npy", sorting.get_property('gt_precision'))
    np.save(path / "gt_recall.npy", sorting.get_property('gt_recall'))


def analyze_sortings(analyses_name: list[str]) -> dict[str, si.BaseSorting]:
    """
    Given a list of analyses name, returns a dictionary of processed sortings.
    The sorting has units correctly aligned, with duplicated spikes removed,
    empty units are also removed, and SNR / contamination are pre-computed.
    GT metrics (comparison with GT sorting) is also computed and stored.
    """

    # STEP 1: Loading and processing sortings
    sortings = {}
    censored_period = COINCIDENT_WINDOW * 1e3 / recording.sampling_frequency

    print("Loading sortings ...")
    for analysis_name in tqdm(analyses_name):
        if (path := analyses_folder / analysis_name / "processed_sorting").exists():
            sortings[analysis_name] = si.load_extractor(path)
            continue
        else:
            print(f"Sorting isn't processed for {analysis_name}")
            print(analyses_folder / analysis_name / "processed_sorting")

        # We need to load and process the sorting.
        sorting = si.load_extractor(analyses_folder / analysis_name / "sorting")
        sorting.annotate(name=analysis_name)
        sorting = sorting.remove_empty_units()
        sorting = scur.remove_duplicated_spikes(sorting, censored_period_ms=censored_period, method="keep_first_iterative")
    
        if len(sorting.to_spike_vector()) > 0:
            # Extracting the templates.
            analyzer = si.create_sorting_analyzer(sorting, recording_f, format="memory", sparse=False)
            analyzer.compute({
                'random_spikes': {'max_spikes_per_unit': params_wvf['max_spikes_per_unit']},
                'templates': {'ms_before': params_wvf['ms_before'], 'ms_after': params_wvf['ms_after']},
                'noise_levels': {},
                'spike_amplitudes': {}
            })
            templates_ext = analyzer.get_extension("templates")
            templates = templates_ext.get_data()
            center = templates_ext.nbefore

            # Shift the templates to put the extremum at t=0
            shifts = {}
            for unit_ind, unit_id in enumerate(sorting.unit_ids):
                template = templates[unit_ind]
                t = np.argmin(np.min(template, axis=1))
                shifts[unit_id] = center - t

            sorting = spost.align_sorting(sorting, shifts)

            # Pre-computing units estimated contamination.
            contaminations, _ = sqm.compute_refrac_period_violations(analyzer, refractory_period_ms=1.9, censored_period_ms=censored_period)
            sorting.set_property("estimated_contamination", list(contaminations.values()))
            
            # Pre-computing SD ratio.
            sd_ratio = sqm.compute_sd_ratio(analyzer)
            sorting.set_property("sd_ratio", list(sd_ratio.values()))

            # Pre-computing the SNR
            SNR = sqm.compute_snrs(analyzer, peak_sign="neg", peak_mode="extremum")
            sorting.set_property("SNR", list(SNR.values()))

            del analyzer
            
        # Saving the processed sorting.
        sortings[analysis_name] = sorting.save(folder=path)


    # STEP 2: Getting ground truth data
    n_spikes_true = gt_sorting.count_num_spikes_per_unit(outputs="array")[None, :]
    gt_matrices = {}
    keys = ['coincidence', 'accuracy', 'recall', 'precision']
    
    print("Calculating coincidence with ground truth ...")
    for analysis_name in tqdm(analyses_name):
        sorting = sortings[analysis_name]

        n_spikes = sorting.count_num_spikes_per_unit(outputs="array")[:, None]
        coincidence_matrix = compute_coincidence_matrix(sorting, gt_sorting)
        
        p = (2 * COINCIDENT_WINDOW + 1) / recording.get_num_frames()
        tp = (coincidence_matrix - n_spikes*n_spikes_true*p) / (1 - n_spikes_true*p)
        fp = n_spikes - tp
        fn = n_spikes_true - tp

        arr_dtype = np.dtype([(key, np.int32 if key=="coincidence" else np.float32, coincidence_matrix.shape[1]) for key in keys])
        arr = np.empty(coincidence_matrix.shape[0], dtype=arr_dtype)

        arr['coincidence'] = coincidence_matrix
        arr['accuracy'] = tp / (tp+fp+fn)
        arr['recall'] = tp / n_spikes_true
        arr['precision'] = tp / n_spikes

        gt_matrices[analysis_name] = arr


    # STEP 3: Getting agreement data
    agreement_matrices = {}

    print("Calculating coincidence with self ...")
    for analysis_name in tqdm(analyses_name):
        sorting = sortings[analysis_name]

        n_spikes = sorting.count_num_spikes_per_unit(outputs="array")
        coincidence_matrix = compute_coincidence_matrix(sorting, sorting)

        agreement_matrices[analysis_name] = coincidence_matrix / (n_spikes[:, None] + n_spikes[None, :] - coincidence_matrix)


    # STEP 4: Constructing "theoretical maximum"
    if not (maximum_folder := analyses_folder / "theoretical_maximum").exists():
        best_units = {}
        
        print("Constructing \"theoretical maximum\" ...")
        for neuron_index, neuron_id in enumerate(tqdm(gt_sorting.unit_ids)):
            best_accuracy = -1.0
            best_spk_train = None

            for sorting_name in analyses_name:
                if sorting_name.startswith("lussac"):  # Only take spike-sorting results to compute theoretical maximum.
                    continue
                sorting = sortings[sorting_name]

                good_units_mask = gt_matrices[sorting_name]['precision'][:, neuron_index] >= MIN_PRECISION_GOOD_NEURON
                if np.sum(good_units_mask) == 0:
                    continue

                best_unit_index = np.argmax(gt_matrices[sorting_name]['accuracy'][good_units_mask, neuron_index])
                best_unit_id = sorting.unit_ids[good_units_mask][best_unit_index]
                best_unit_index = sorting.id_to_index(best_unit_id)

                if (best_acc := gt_matrices[sorting_name]['accuracy'][best_unit_index, neuron_index]) > best_accuracy:
                    best_accuracy = best_acc
                    best_spk_train = sorting.get_unit_spike_train(best_unit_id)

            if best_spk_train is not None:
                best_units[neuron_id] = best_spk_train

        sorting = si.NumpySorting.from_unit_dict(best_units, recording.sampling_frequency)
        sorting.save(folder = maximum_folder / "sorting")


    # STEP 5: Categorizing units
    print("Categorizing units ...")
    for analysis_name in tqdm(analyses_name):
        sorting = sortings[analysis_name]
        unit_ids = sorting.unit_ids
        
        if len(unit_ids) == 0:
            continue
        
        if 'gt_label' in sorting.get_property_keys():
            continue

        units_category = {}
        units_neuron = {}
        sorting.set_property('gt_accuracy', [0.0]*sorting.get_num_units())
        sorting.set_property('gt_precision', [0.0]*sorting.get_num_units())
        sorting.set_property('gt_recall', [0.0]*sorting.get_num_units())

        # Categorizing small units
        firing_rates = sorting.count_num_spikes_per_unit(outputs="array") / recording.get_num_samples() * recording.sampling_frequency
        small_units = unit_ids[firing_rates < SMALL_UNITS_MAX_FR]
        units_category = {unit_id: "Small" for unit_id in small_units} | units_category

        # Categorizing contaminated units
        contamination = sorting.get_property('estimated_contamination')
        contaminated_units = unit_ids[contamination > MAX_CONTAMINATION]
        units_category = {unit_id: "Contaminated" for unit_id in contaminated_units} | units_category

        # Categorizing noisy units
        fraction_spikes = np.sum(gt_matrices[analysis_name]['precision'], axis=1)
        noisy_units = unit_ids[fraction_spikes < MIN_FRAC_SPIKES_NOISY]
        units_category = {unit_id: "Noise" for unit_id in noisy_units} | units_category

        # Categorizing merged units
        recall_2nd_best = np.sort(gt_matrices[analysis_name]['recall'], axis=1)[:, -2]
        merged_units = unit_ids[recall_2nd_best > RECALL_2ND_BEST_MERGED]
        units_category = {unit_id: "Merged" for unit_id in merged_units} | units_category

        # Categorizing multi-unit activity
        precision_best = np.max(gt_matrices[analysis_name]['precision'], axis=1)
        mua_units = unit_ids[precision_best < MUA_MAX_PRECISION]
        units_category = {unit_id: "MUA" for unit_id in mua_units} | units_category

        # Now, each unit not categorized yet needs to be matched to a neuron
        accuracies = gt_matrices[analysis_name]['accuracy']
        matching_neuron = np.argmax(accuracies, axis=1)  # For each unit, the neuron_index that best matches.

        for neuron_index, neuron_id in enumerate(gt_sorting.unit_ids):
            unit_indices = np.argwhere(matching_neuron == neuron_index)[:, 0]
            unit_indices = [unit_index for unit_index in unit_indices if unit_ids[unit_index] not in units_category]
            best_unit_index = np.argmax(accuracies[:, neuron_index])

            for unit_index in unit_indices:
                unit_id = unit_ids[unit_index]
                units_neuron[unit_id] = neuron_id

                sorting._properties['gt_accuracy'][unit_index] = gt_matrices[analysis_name]['accuracy'][unit_index, neuron_index]
                sorting._properties['gt_precision'][unit_index] = gt_matrices[analysis_name]['precision'][unit_index, neuron_index]
                sorting._properties['gt_recall'][unit_index] = gt_matrices[analysis_name]['recall'][unit_index, neuron_index]

                if unit_index == best_unit_index:
                    accuracy = accuracies[unit_index, neuron_index]
                    for i in range(len(MATCH_ACC)):
                        if accuracy >= MATCH_ACC[i]:
                            quality = MATCH_LABELS[i]
                            break
                    else:
                        quality = MATCH_LABELS[-1]
                    units_category[unit_id] = quality
                elif agreement_matrices[analysis_name][unit_index, best_unit_index] >= SHARED_FRAC_REDUNDANT:
                    units_category[unit_id] = "Redundant"
                else:
                    units_category[unit_id] = "Split"

        # Saving properties in the sorting
        sorting.set_property('gt_label', list(units_category.values()), ids=list(units_category.keys()))
        sorting.set_property('gt_neuron', list(units_neuron.values()), ids=list(units_neuron.keys()))
        save_gt_properties(sorting, analyses_folder / analysis_name / "processed_sorting" / "properties")
        
        # Also saving the properties in the original sorting
        original_sorting = si.load_extractor(analyses_folder / analysis_name / "sorting")
        for key in ["gt_label", "gt_neuron", "gt_accuracy", "gt_precision", "gt_recall"]: 
            original_sorting.set_property(key, sorting.get_property(key), ids=sorting.unit_ids)
        save_gt_properties(original_sorting, analyses_folder / analysis_name / "sorting" / "properties")
    
    return sortings


sortings = analyze_sortings(list(analyses_params.keys()))
```

```python
ppi = 300
px_per_point = 1/72 * ppi
width_cm = 16.0
height_cm = 8.0
width_in = width_cm / 2.54
height_in = height_cm / 2.54
width = int(round(width_in * ppi))
height = int(round(height_in * ppi))

def plot_units_categorization(sortings: dict[str, si.BaseSorting]) -> None:
    names = list(sortings.keys())
    results = {}
    
    for analysis_name in names:
        # Count the number of units with each category for the results
        labels, counts = np.unique(sortings[analysis_name].get_property('gt_label'), return_counts=True)
        results[analysis_name] = dict(zip(labels, counts))
    
        # If no units have the category, then we need to set it to zero (to prevent KeyError)
        keys = [*MATCH_LABELS, 'Merged', 'Split', 'Redundant', 'MUA', 'Noise']
        for key in keys:
            if key not in results[analysis_name]:
                results[analysis_name][key] = 0

    matched_df = pd.DataFrame(
        [[results[name][MATCH_LABELS[0]], results[name][MATCH_LABELS[1]], -results[name][MATCH_LABELS[2]], -results[name][MATCH_LABELS[3]]] for name in names],
        index=names,
        columns=MATCH_LABELS
    )
    split_df = pd.DataFrame(
        [[-results[name]['Split'], -results[name]['Redundant']] for name in names],
        index=names,
        columns=['Split', 'Redundant']
    )
    merged_df = pd.DataFrame(
        [[-results[name]['Merged']] for name in names],
        index=names,
        columns=['Merged']
    )
    noisy_df = pd.DataFrame(
        [[-results[name]['MUA'], -results[name]['Noise']] for name in names],
        index=names,
        columns=['MUA', 'Noise']
    )

    df = pd.concat([matched_df, split_df, merged_df, noisy_df], axis=1, keys=["Matched", "Under-aggregated", "Over-aggregated", "Garbage"])
    
    colors = {
        'Matched': {MATCH_LABELS[0]: "#006400", MATCH_LABELS[1]: "#228B22", MATCH_LABELS[2]: "#90EE90", MATCH_LABELS[3]: "#ACE3D2"},
        'Under-aggregated': {'Split': "#6495ED", 'Redundant': "#008B8B"},
        'Over-aggregated': {'Merged': "#DC143C"},
        'Garbage': {'MUA': "#BA55D3", 'Noise': "#000000"}
    }

    fig = go.Figure()

    for i, t in enumerate(colors):
        for j, col in enumerate(df[t].columns):
            fig.add_trace(go.Bar(
                x=[name.replace('_', "<br />") for name in df.index],
                y=df[t][col],
                yaxis=f"y{i+1}",
                offsetgroup=str(i),
                offset=(i-1)*0.15,
                width=0.15,
                legendgroup=t,
                legendgrouptitle_text=t,
                name=col,
                marker_color=colors[t][col],
                hovertemplate="%{y}<extra></extra>"
            ))

    fig.update_layout(go.Layout(
        barmode="relative",
        yaxis_showticklabels=False,
        yaxis_showgrid=False,
        yaxis2=go.layout.YAxis(
            visible=False,
            matches='y',
            overlaying='y',
            anchor='x'
        ),
        yaxis3=go.layout.YAxis(
            visible=False,
            matches='y',
            overlaying='y',
            anchor='x'
        ),
        yaxis4=go.layout.YAxis(
            visible=False,
            matches='y',
            overlaying='y',
            anchor='x'
        ),
        legend=dict(x=0.99, y=0.99, xanchor="right", yanchor="top", orientation='h', bgcolor="rgba(0, 0, 0, 0.0)"),
        hovermode='x',
        margin=dict(b=5, t=10, l=5, r=10)
    ))

    fig.show(renderer="svg")
    # fig.show(renderer="browser")


plot_units_categorization(sortings)
```

```python
def plot_accuracy_by_SNR(sortings: dict[str, si.BaseSorting]) -> None:
    df = pd.DataFrame(0, index=gt_sorting.unit_ids, columns=["SNR", *sortings.keys()], dtype=np.float32)

    df['SNR'] = gt_sorting.get_property('SNR')

    for analysis_name in sortings.keys():
        sorting = sortings[analysis_name]
        
        for unit_id in sorting.unit_ids:
            neuron_id = sorting.get_unit_property(unit_id, 'gt_neuron')
            if neuron_id == "":
                continue

            accuracy = sorting.get_unit_property(unit_id, 'gt_accuracy')
#             accuracy = 0 if accuracy < 0.8 else accuracy
            df.loc[neuron_id, analysis_name] = max(df[analysis_name][neuron_id], accuracy)

    dx = 0.05
    SNR_axis = np.arange(0, df['SNR'].max() + dx, dx)
    SNR_std = 0.4
    SNR_gaussians = np.exp(-(np.array(df['SNR'], dtype=np.float64)[:, None] - SNR_axis[None, :])**2 / (2 * SNR_std**2))

    fig = go.Figure()

    for i, analysis_name in enumerate(sortings.keys()):
        accuracies = np.sum(np.array(df[analysis_name])[:, None] * SNR_gaussians, axis=0) / np.sum(SNR_gaussians, axis=0)

        # Plot mean
        fig.add_trace(go.Scatter(
            x=SNR_axis,
            y=accuracies,
            mode="lines",
            marker_color=plotly.colors.qualitative.Alphabet[i],
            name=analysis_name
        ))

        # Plot individual points
        fig.add_trace(go.Scatter(
            x=df['SNR'],
            y=df[analysis_name],
            mode="markers",
            marker_opacity=0.7,
            marker_color=plotly.colors.qualitative.Alphabet[i],
            name=f"{analysis_name} data",
            text=gt_sorting.unit_ids,
            visible="legendonly"
        ))

    fig.update_xaxes(title_text="SNR (Signal-to-Noise Ratio)", type="log", range=[math.log10(0.7), math.log10(df['SNR'].max())])
    fig.update_yaxes(title_text="Averaged accuracy")
    fig.show(renderer="svg")


plot_accuracy_by_SNR(sortings)
```

```python
def plot_propotion_by_accuracy(sortings: dict[str, si.BaseSorting]) -> None:
    df = pd.DataFrame(0, index=gt_sorting.unit_ids, columns=list(sortings.keys()), dtype=np.float32)
    
    for analysis_name in sortings.keys():
        sorting = sortings[analysis_name]
        
        for unit_id in sorting.unit_ids:
            neuron_id = sorting.get_unit_property(unit_id, 'gt_neuron')
            if neuron_id == "":
                continue
            accuracy = sorting.get_unit_property(unit_id, 'gt_accuracy')
            df.loc[neuron_id, analysis_name] = max(df[analysis_name][neuron_id], accuracy)
    
    dx = 0.0005
    xaxis = np.arange(0, 1+dx/2, dx)
    
    fig = go.Figure()
    
    for i, analysis_name in enumerate(sortings.keys()):
        fig.add_trace(go.Scatter(
            x=xaxis,
            y=np.sum(df[analysis_name].to_numpy()[:, None] >= xaxis[None, :], axis=0),
            mode="lines",
            name=analysis_name
        ))

    fig.update_xaxes(title_text="Accuracy", range=[0.7, 1.0])
    fig.update_yaxes(title_text="Number of neurons", range=[-20, 400])
    fig.show(renderer="svg")
    # fig.show(renderer="browser")
    

plot_propotion_by_accuracy(sortings)
```
